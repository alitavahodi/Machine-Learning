```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 15, 100)
y = np.sin(x) + 0.1 * np.random.randn(len(x))

plt.figure(figsize=(10, 4))

plt.plot(x, y, 'o--', color='purple', lw=1.3, ms=5, label='test')

plt.legend(loc='lower right', fontsize=6)

plt.show()
```

```python
y_1 = np.sin(x)

plt.figure(figsize=(10, 4))

plt.plot(x, y, 'o', color='purple', ms=5, label='Data')
plt.plot(x, y_1, color='green', label='Theory')

plt.xlabel('Time [s]')
plt.ylabel('Voltage [V]')

# plt.xlim()
plt.ylim(top=1.4)

plt.legend(loc='upper right', ncol=10, fontsize=10)

plt.savefig('asfk.png', dpi=300)

plt.show()
```

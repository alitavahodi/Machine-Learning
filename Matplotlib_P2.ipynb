```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gas = pd.read_csv("gas_prices.csv")
gas
```

```python
plt.figure(figsize=(9, 5))

# plt.plot(gas.Year, gas.Canada, label='Canada')
# plt.plot(gas.Year, gas.Germany, label='Germany')
# plt.plot(gas.Year, gas.France, label='France')
# plt.plot(gas.Year, gas.Mexico, label='Mexico')
# plt.plot(gas.Year, gas.Japan, label='Japan')

countries = ['Canada', 'Germany', 'France', 'Mexico', 'Japan']

for country in countries:
    plt.plot(gas.Year, gas[country],  '--o', label=country)

plt.title("Gas Prices over time (in USD)", fontdict={"fontweight": 'bold', 'fontsize': 20})

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.xticks(gas.Year[::3])

plt.legend()

plt.savefig('gas_prices.png', dpi=300)

plt.show()
```

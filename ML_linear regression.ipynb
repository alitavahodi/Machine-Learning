```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import csv
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv")
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df2=df[df["Position"]=='GK']
df2.sort_values(by="Overall")
df2.reset_index(inplace=True)
df2.head()
```

```python
cdf=df2[['Overall','Jumping','GKDiving','GKPositioning']]

cdf.head(10)
```

```python
cdf.dtypes
cdf
```

```python
cdf.iloc[550]
```

```python
#Removing a table values and input new values in pandas
cdf
```

```python
viz=cdf[['Overall','Jumping','GKDiving','GKPositioning']]
viz.hist()
plt.show()
```

```python
fig, axes = plt.subplots(3, figsize=(5, 15))
ax=axes [0]
ax.scatter(cdf.Jumping,cdf.Overall, color='blue')
ax.set_xlabel('Jumping')
ax.set_ylabel('Overall')

ax=axes[1]
ax.scatter(cdf.GKDiving,cdf.Overall, color='blue')
ax.set_xlabel('GKDiving')
ax.set_ylabel('Overall')

ax=axes[2]
ax.scatter(cdf.GKPositioning,cdf.Overall, color='blue')
ax.set_xlabel('GKPositioning')
ax.set_ylabel('Overall')
plt.show()
```

```python
msk=np.random.rand(len(df2))<0.8
train=cdf[msk]
test=cdf[~msk]
```

```python
train
```

```python
test
```

# Predicting Overal based on GKDiving

```python
fig, ax = plt.subplots(1)

ax.scatter(train.GKDiving,train.Overall, color='gray')
ax.scatter(test.GKDiving,test.Overall, color='red')
ax.set_xlabel('GKDiving')
ax.set_ylabel('Overall')
plt.show()
```

```python
from sklearn import linear_model
regr=linear_model.LinearRegression()
train_x=np.asanyarray(train[['GKDiving']])
train_y=np.asanyarray(train[['Overall']])
regr.fit(train_x,train_y)
print('Coefficients: ', regr.coef_)
print('Intercept: ', regr.intercept_)
```

```python
plt.scatter(train.GKDiving,train.Overall, color='gray')
plt.plot(train_x,regr.coef_[0][0]*train_x+regr.intercept_[0], '-r')
plt.xlabel('GKDiving')
plt.ylabel('Overall')
```

```python
from sklearn.metrics import r2_score

test_x=np.asanyarray(test[["GKDiving"]])
test_y=np.asanyarray(test[["Overall"]])
test_y_=regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_-test_y)))
print("Residual sum of squares (MSE): %.2F" %np.mean((test_y_-test_y)**2))
print("R2-score: %.2f" % r2_score(test_y,test_y_))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 2, 4, 6, 8, 10]

plt.figure(figsize=(10, 6))

# line 2x
plt.plot(x, y, 'b.--', ms=10, label='2x')

# x**2
x_1 = np.arange(0, 4.5, 0.5)

plt.plot(x_1[:6], x_1[:6]**2, 'r', label='x^2')
plt.plot(x_1[5:], x_1[5:]**2, 'r--')

plt.title("Our First Graph", fontdict={'fontname': 'Arial', 'fontsize': 20})

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.xticks([0, 1, 2, 3, 4, 5, 6])
# plt.yticks()

plt.grid(False)

plt.legend(loc='upper left', fontsize=20)

plt.savefig('first_graph.png', dpi=300)

plt.show()
```

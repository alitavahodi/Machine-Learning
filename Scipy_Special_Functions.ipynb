## Special Functions

**(A)** Compute Bessel function of order 3 at x = 5

```python
import scipy.special
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 5.0
n = 3
result = scipy.special.jn(n, x)
print(f"J_{n}({x}) =", result)

x_values = np.linspace(0, 10, 100)

n = 3
bessel_values = scipy.special.jn(n, x_values)
plt.plot(x_values, bessel_values)
plt.title(f"Bessel function J_{n}(x)")
plt.xlabel("x")
plt.ylabel(f"J_{n}(x)")
plt.grid(True)
plt.show()
```

**(B)** Compute gamma function for x=7 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special

x = 7.0
result_gamma = scipy.special.gamma(x)
print("Gamma(7) =", result_gamma)

x_values = np.linspace(0.1, 7, 100)
gamma_values = scipy.special.gamma(x_values)
plt.plot(x_values, gamma_values)
plt.title("Gamma Function")
plt.xlabel("x")
plt.ylabel("Gamma(x)")
plt.grid(True)
plt.show()
```

**(C)** Legendre polynomials are orthogonal over the interval [-1, 1] with weight function 1. Plot the first 5 Legendre polynomials.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special

x = np.linspace(-1,1, 100)
for n in range(5):
    legendre_poly = scipy.special.legendre(n)
    plt.plot(x, legendre_poly(x), label=f'Legendre P_{n}(x)')
plt.title('Legendre Polynomials')
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.legend()
plt.grid(True)
plt.show()
```

**(D)** plot sin(x) and cos(x) along the interval $[0,530]$ in degrees

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special

angles_degrees = np.linspace(0, 530, 100)


sine_values = scipy.special.sindg(angles_degrees)
cosine_values = scipy.special.cosdg(angles_degrees)


plt.plot(angles_degrees, sine_values, label='Sine')
plt.plot(angles_degrees, cosine_values, label='Cosine')
plt.xlabel('Angle (degrees)')
plt.ylabel('Value')
plt.title('Sine and Cosine Functions')
plt.legend()
plt.grid(True)
plt.show()
```
